```python
def find_str_len(s):
    if type(s) == str:
        count = 0
        for i in s:
            count+=1
        return count
    else:
        raise ValueError(f'{s} is not a sring')

find_str_len("This is a String")
    
# find_str_len(15)
```

### 2

```python
def get_index(l,i):
    ind = 0
    for val in l:
        if val == i:
            return ind
        ind += 1
    raise ValueError(f'{i} is not in list')

l = [0,1,2,3,4,5]
get_index(l,5)
```

```python
def get_sys_ip():
    import socket as sc
    hostname = sc.gethostname()
    ip_add = sc.gethostbyname(hostname)
    print(f"IP of your system is {ip_add}")
    return ip_add
get_sys_ip()
```

```python
def can_convert_float(n):
    try:
        float(n)
        return True
    except ValueError:
        return False
    
def product_of_num(l):
    product = 1
    flag = True
    for i in l:
        if can_convert_float(i):
            flag = False
            product *= float(i)
    if flag:
        raise ValueError("Any numeric value not found in list")
    return product

l = [3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']
product_of_num(l)
```

```python
import smtplib
import time
import imaplib
import email
import traceback 


ORG_EMAIL = "@gmail.com" 
FROM_EMAIL = "kj3.2468" + ORG_EMAIL 
FROM_PWD = "zvurfptnugqmjsor" 
SMTP_SERVER = "imap.gmail.com" 
SMTP_PORT = 993

def read_email_from_gmail(ORG_EMAIL,FROM_EMAIL,FROM_PWD,SMTP_SERVER,SMTP_PORT):
    try:
        mail = imaplib.IMAP4_SSL(SMTP_SERVER)
        mail.login(FROM_EMAIL,FROM_PWD)
        mail.select('inbox')

        data = mail.search(None, 'ALL')
        mail_ids = data[1]
        id_list = mail_ids[0].split()   
        first_email_id = int(id_list[0])
        latest_email_id = int(id_list[-1])

        for i in range(latest_email_id,first_email_id, -1):
            data = mail.fetch(str(i), '(RFC822)' )
            for response_part in data:
                arr = response_part[0]
                if isinstance(arr, tuple):
                    msg = email.message_from_string(str(arr[1],'utf-8'))
                    email_subject = msg['subject']
                    email_from = msg['from']
                    print('From : ' + email_from + '\n')
                    print('Subject : ' + email_subject + '\n')

    except Exception as e:
        traceback.print_exc() 
        print(str(e))
        

read_email_from_gmail(ORG_EMAIL,FROM_EMAIL,FROM_PWD,SMTP_SERVER,SMTP_PORT)
```

```python
import smtplib
import email
import traceback 

def send_mails(sender_email,sender_pass,receiver_email,msg_body):
    
    host_name = ((sender_email).split('@'))[1]
    if(host_name == 'outlook.com'):
        host_name = "smtp-mail.outlook.com"
    else:
        if(host_name == 'att.net' or host_name == 'yahoo.com'):
            host_name = "mail." + host_name
        elif(host_name == 'icloud.com/mail'):
            host_name = 'mail.me.com'
        host_name = "smtp." + host_name
    try:
        server = smtplib.SMTP(host = host_name,port = 587)
    except:
        server = smtplib.SMTP(host = host_name,port = 2525)
          
    try:
        server.starttls()
        server.login(sender_email,sender_pass)
    except Exception as e:
            traceback.print_exc() 
            print(str(e))
    

    from email.mime.multipart import MIMEMultipart
    from email.mime.text import MIMEText

    msg = MIMEMultipart()
    message = msg_body

    msg['From']=sender_email
    msg['To']=receiver_email
    
    msg['Subject']="Python function to send mail"
    msg.attach(MIMEText(message,'plain'))


    server.send_message(msg)

    server.quit()

sender_email = "kj3.2468@gmail.com"
sender_pass = "zvurfptnugqmjsor"
receiver_email = "kushjoshi16@gmail.com"
msg_body = "This is a custom msg sent using python function"

send_mails(sender_email,sender_pass,receiver_email,msg_body)
```

# Multivariate Analysis
### What are the Top 5 most installed Apps in Each popular Categories ??

```python
df.head()
```

install--ap--category

```python
dfApps = df.groupby(['Category','App'])['Installs'].sum().reset_index()
```

```python
dfApps
```

```python
dfa = dfApps.sort_values('Installs',ascending=False)
```

```python
dfa[dfa.Category=="GAME"][:5]
```

```python
# def get_popularCat():
#     for i in df2['Category']:
#         yield i
```

```python
# popCat = get_popularCat()
```

```python
# print(next(popCat))
```

```python
# dfa[dfa.Category==next(popCat)][:5]
```

```python
def get_popularCat():
    for i in df2['Category'][:8]:
        yield i

popCat = get_popularCat()

sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))
for i,app in enumerate(popCat):
#     print(i,app)
    dfg2 = dfa[dfa.Category == app]
    dfg3 = dfg2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data = dfg3, x= 'Installs',y='App')
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size=20)

plt.tight_layout()
plt.subplots_adjust(hspace = .3)
plt.show()
```

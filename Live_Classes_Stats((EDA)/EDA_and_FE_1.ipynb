```python
# data_copy = data_copy["Size"].str.replace("M","000k")
# data_copy = data_copy["Size"].str.replace("k","")

def size_conv(size):
    if size[-1] == 'M':
        return float(size[:-1])*1024
    elif size[-1] == 'k':
        return float(size[:-1])
    else:
        return float('nan')

def size_convMod(size):
    match(size[-1]):
        case 'M':
            return float(size[:-1])*1024
        case 'k':
            return float(size[:-1])
        case _:
            return float('nan')
```

```python
data_copy['Size'] = data_copy['Size'].apply(size_convMod)
```
